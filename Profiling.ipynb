# Report on Market Share for all the vehicle types

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(y= ev_data['vehicle type'],width=ev_data['market share'])
ax.set_title("Segmentation for Market Share")
ax.set_xlabel("Market share");
```

##### Based on the above graph it is clear that the market share for "M-CYCLE/SCOOTER" and "E-RICKSHAW" is high compared to all the vehicles. But to enter a market we need a good market share for a vehicle in the above graph it is clear that the "Motor Car" has a good share which can help us to get a good market.

# Max sales

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(y= ev_data['vehicle type'],width=ev_data['max sales volume'])
ax.set_title("Segmentation for Maximum sales")
ax.set_xlabel("Max Sales Volume");
```

###### For a company to enter the E-Rickshaw and M-cycle/scooter market is really a big challenge has it requires more invesment. # Place to choose for the Production and Manufacturing.

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(10,5))
ax1.barh(y= ev_data['state'],width=ev_data['Public EV Charging Stations'])
ax1.set_title("Segmentation for States with EV Charging Station")
ax1.set_xlabel("Public EV Charging Stations")
ax2.barh(y= ev_data['state'],width=ev_data['Average gross income '])
for tick in ax2.get_xticklabels():
    tick.set_rotation(90)
ax2.set_title("Segmentation for States with Average Gross Income")
ax2.set_xlabel("Average gross income")
fig.tight_layout();
```

##### Based on the graph it is clear that, "DELHI" can be choosed as the base for production and manufacturing has it has high EV Charging Station and high Average gross income and apart from DELHI the next place would be MAHARASTRA because of the high number in EV Charging Station and Average gross income.
